```python
"""
car_id = 'А222BС96'
Результат: Номер А222BС валиден. Регион: 96
car_id = 'АБ22ВВ193'
Результат: Номер не валиден
"""
import re

number = ['А222BС96', 'АБ22ВВ193']

for id in number:
    if id in re.findall(r'\w\d{3}\w{2}\d{2,3}', id):
        print('Номер {} валиден. Регион: {}'.format(id, id[-(len(id)-6):]))
    else:
        print('Номер {} не валиден.'.format(id))
```
